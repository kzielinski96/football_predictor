## Files load

```python
import os
import pandas as pd

def load_football_data(football_path, file):
    csv_path = os.path.join(football_path, file)
    return pd.read_csv(csv_path, error_bad_lines=False)
```

```python
football = load_football_data(FOOTBALL_PATH_GE, "D1 (19_20).csv")
```
